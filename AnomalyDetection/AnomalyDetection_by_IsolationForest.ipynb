```python
# To create individual dataframes for each business and store them in a dictionary:
business_dict = {name: biz_df[biz_df['Third Party Name'] == name] for name in biz_df['Third Party Name'].unique()}

# Let's print out the names of the businesses to confirm creation of the dictionary
list(business_dict.keys())[:10]  # Display only the first 10 for brevity
```

```python
from sklearn.ensemble import IsolationForest

# Select numerical columns for anomaly detection
numerical_data = biz_df[['Balance', 'Amount']]

# Initialize the isolation forest model
iforest = IsolationForest(n_estimators=100, contamination=0.05) 
iforest.fit(numerical_data)

# Predicting anomalies (outliers)
anomalies = iforest.predict(numerical_data)
anomaly_indices = anomalies == -1 

# Add a column for anomalies to the original dataset
biz_df['Anomaly'] = anomaly_indices

# Count anomalies and display some of them
anomaly_count = anomaly_indices.sum()
anomaly_examples = biz_df[biz_df['Anomaly']].head()
```
