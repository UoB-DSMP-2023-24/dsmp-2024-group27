## Feature engineering for anomaly detection

### Calculating general thresholds using Q3 and IQR and a multiple

```python
# Calculating Q1, Q3, and IQR for each account in 'from_totally_fake_account'
quartiles = original_df.groupby('from_totally_fake_account')['monopoly_money_amount'].quantile([0.25, 0.75]).unstack()
quartiles['IQR'] = quartiles[0.75] - quartiles[0.25]

# Calculate thresholds for overall spending for user across all industries/users based on IQR
thresholds = quartiles.copy()
thresholds['Overall_Threshold_1_IQR'] = quartiles[0.75] + 2 * quartiles['IQR']
thresholds['Overall_Threshold_2_IQR'] = quartiles[0.75] + 3 * quartiles['IQR']
thresholds['Overall_Threshold_3_IQR'] = quartiles[0.75] + 4 * quartiles['IQR']

# Creating a new dataframe for anomaly detection with threshold columns
anomaly_detection_df = pd.merge(user_df_updated[['user_id']], thresholds[['Overall_Threshold_1_IQR', 'Overall_Threshold_2_IQR', 'Overall_Threshold_3_IQR']], how='left', left_on='user_id', right_index=True)

# Display the newly created anomaly_detection_df DataFrame
anomaly_detection_df.head()
```

```python
# Convert 'not_happened_yet_date' to datetime format, ensuring dayfirst format
original_df['not_happened_yet_date'] = pd.to_datetime(original_df['not_happened_yet_date'], dayfirst=True)

# Calculate the median number of transactions per day, per week, and per month per user
user_daily_trans = original_df.groupby(['from_totally_fake_account', original_df['not_happened_yet_date'].dt.date]).size().groupby(level=0).median()
user_weekly_trans = original_df.groupby(['from_totally_fake_account', original_df['not_happened_yet_date'].dt.isocalendar().week]).size().groupby(level=0).median()
user_monthly_trans = original_df.groupby(['from_totally_fake_account', original_df['not_happened_yet_date'].dt.month]).size().groupby(level=0).median()

# Add calculated medians to the anomaly_detection_df DataFrame
anomaly_detection_df['median_trans_per_day'] = anomaly_detection_df['user_id'].map(user_daily_trans)
anomaly_detection_df['median_trans_per_week'] = anomaly_detection_df['user_id'].map(user_weekly_trans)
anomaly_detection_df['median_trans_per_month'] = anomaly_detection_df['user_id'].map(user_monthly_trans)

anomaly_detection_df.head()
```

### Creating a user and business account df

```python
original_df['to_randomly_generated_account'] = original_df['to_randomly_generated_account'].astype(str)

# Split the DataFrame into user accounts and business accounts without creating a new column
user_accounts_df = original_df[original_df['to_randomly_generated_account'].str.isdigit()]
business_accounts_df = original_df[~original_df['to_randomly_generated_account'].str.isdigit()]

user_accounts_df.shape, business_accounts_df.shape
```

```python
# Group by 'from_totally_fake_account' and aggregate the 'to_randomly_generated_account' into a list
user_to_user_transfers = user_accounts_df.groupby('from_totally_fake_account')['to_randomly_generated_account'].apply(list)

# Calculate the Q3 + 1.5*IQR for each user to each account they transfer to
def calculate_thresholds(group):
    # Calculate Q1, Q3, and IQR for each 'to_randomly_generated_account'
    quantiles = group['monopoly_money_amount'].quantile([0.25, 0.75])
    IQR = quantiles[0.75] - quantiles[0.25]
    threshold = quantiles[0.75] + 1.5 * IQR
    return threshold

# Group by 'from_totally_fake_account' and 'to_randomly_generated_account', then apply the threshold calculation
transfer_thresholds = user_accounts_df.groupby(['from_totally_fake_account', 'to_randomly_generated_account']).apply(calculate_thresholds).reset_index()

# Pivot the results to get a list of thresholds for each 'from_totally_fake_account'
pivot_transfer_thresholds = transfer_thresholds.pivot(index='from_totally_fake_account', columns='to_randomly_generated_account', values=0)

# Convert the pivot table to a dictionary {user_account: [list_of_thresholds]}
thresholds_dict = pivot_transfer_thresholds.apply(lambda row: row.dropna().to_list(), axis=1).to_dict()

# Add the list of accounts each user transfers to and the corresponding thresholds to the original DataFrame
user_accounts_df['transferred_to_accounts'] = user_accounts_df['from_totally_fake_account'].map(user_to_user_transfers)
user_accounts_df['transfer_thresholds'] = user_accounts_df['from_totally_fake_account'].map(thresholds_dict)

user_accounts_df.head(30)
```
