### Loading the dataset

```python
# Loading csv and setting to 
user_df = pd.read_csv('../SavedData/Lloyds_users.csv')
original_df= pd.read_csv('../fake_transactional_data_24.csv')


# Changing the lists back to normal columns
user_df['user_biz_interaction'] = user_df['user_biz_interaction'].apply(ast.literal_eval)
user_df['user_biz_interaction_frequency'] = user_df['user_biz_interaction_frequency'].apply(ast.literal_eval)
user_df['user_user_interaction'] = user_df['user_user_interaction'].apply(ast.literal_eval)
user_df['user_user_interaction_frequency'] = user_df['user_user_interaction_frequency'].apply(ast.literal_eval)
user_df['user_industry_interaction'] = user_df['user_industry_interaction'].apply(ast.literal_eval)
user_df['user_industry_interaction_frequency'] = user_df['user_industry_interaction_frequency'].apply(ast.literal_eval)
user_df['user_date_transaction'] = user_df['user_date_transaction'].apply(ast.literal_eval)
user_df['user_date_transaction_freq'] = user_df['user_date_transaction_freq'].apply(ast.literal_eval)
```

```python
import pandas as pd

# Calculate Q1, Q3, and IQR for each account in 'from_totally_fake_account'
quartiles = original_df.groupby('from_totally_fake_account')['monopoly_money_amount'].quantile([0.25, 0.75]).unstack()
quartiles.columns = ['Q1', 'Q3']
quartiles['IQR'] = quartiles['Q3'] - quartiles['Q1']


# Prepare quartiles DataFrame for merging by selecting only the necessary columns
quartiles_for_merge = quartiles[['Q1', 'Q3', 'IQR']].copy()

# Merge the prepared quartiles DataFrame with user_df based on 'from_totally_fake_account' without adding it to user_df
# Note: Ensure 'from_totally_fake_account' in original_df matches 'user_id' in user_df for correct merge operation
user_df_updated = pd.merge(user_df_updated, quartiles_for_merge, how='left', left_on='user_id', right_index=True)

# Now user_df will have the columns 'Q1', 'Q3', and 'IQR' for each user account
user_df_updated
```

```python
user_df_updated['ThresholdIQR'] = user_df_updated['Q3'] + 3 * user_df_updated['IQR']
```

```python
import pandas as pd

# Assuming original_df is your transaction data with 'from_totally_fake_account' and 'monopoly_money_amount'

# Step 1: Calculate the mean and standard deviation for each account's transactions
stats = original_df.groupby('from_totally_fake_account')['monopoly_money_amount'].agg(['mean', 'std'])

# Step 2: Define a multiplier for the standard deviation to set the threshold
# For example, using 2 standard deviations from the mean
multiplier = 2

# Step 3: Calculate the threshold for each account
# The threshold is defined as the mean plus the multiplier times the standard deviation
stats['ThresholdSd'] = stats['mean'] + multiplier * stats['std']

# Reset the index to make 'from_totally_fake_account' a column again for merging
stats = stats.reset_index()

# Step 4: Merge the calculated thresholds back into user_df
user_df_updated = pd.merge(user_df_updated, stats[['from_totally_fake_account', 'ThresholdSd']], how='left', left_on='user_id', right_on='from_totally_fake_account')

# Optionally, you can drop the 'from_totally_fake_account' column if it's not needed
user_df_updated.drop(columns=['from_totally_fake_account'], inplace=True)

# Now user_df has a 'Threshold' column for each user account
user_df_updated
```

```python
import pandas as pd

# Assuming original_df is your DataFrame with transaction data
# and includes 'from_totally_fake_account' for the account ID
# and 'monopoly_money_amount' for the transaction amounts

# Step 1: Calculate the mean transaction amount for each account
mean_spending = original_df.groupby('from_totally_fake_account')['monopoly_money_amount'].mean().reset_index()

# Rename the columns for clarity
mean_spending.columns = ['from_totally_fake_account', 'sent_tx_mean_gbp']

# Example merge operation (assuming mean_spending and user_df are already defined)
user_df_updated = pd.merge(user_df_updated, mean_spending, how='left', left_on='user_id', right_on='from_totally_fake_account')

# Drop the 'from_totally_fake_account' column to avoid redundancy
user_df_updated.drop(columns=['from_totally_fake_account'], inplace=True)
```
